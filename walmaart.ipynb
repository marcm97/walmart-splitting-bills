```python
walmart_items =[]
walmart_cost =[]
for order in items:

    if order.find_all("div",{"data-automation-id" :"details-table-header"})[0].text=="Unavailable":
        continue
    
    elif order.find_all("div",{"data-automation-id" :"details-table-header"})[0].text=="Substitutions":
        order_items = order.find_all(class_="OrderItemTile__rowName___8dQXg")[1::2]
        order_prices = order.find_all(class_="OrderItemTile__rowPrice___22B7_")[1::2]
        
    else:
        order_items = order.find_all(class_="OrderItemTile__rowName___8dQXg")
        order_prices = order.find_all(class_="OrderItemTile__rowPrice___22B7_")
        
    for item, price in zip(order_items,order_prices):
        walmart_items.append(item.text)
        walmart_cost.append(float(price.text.split("(")[0][1:]))
        
shipping = html_soup.find_all("div",{"data-automation-id" :"shippingCharges"})[0].text[1:]
tax = html_soup.find_all("div",{"data-automation-id" :"totalTax"})[0].text[1:]
#tip =html_soup.find_all("div",{"data-automation-id" :"driverTipTotal"})[0].text[1:]

for extra in ['shipping','tax']:
    walmart_items.append(str(extra))
    walmart_cost.append(float(eval(extra)))
```

```python
walmart = pd.DataFrame(list(zip(walmart_items, walmart_cost)), 
               columns =['Items', 'Cost'])
```
